# import edilmesi gereken kütüphaneler ve veri setlerini import ettik. 

```python
import pandas as pd
import numpy as np
from sklearn.neighbors import NearestNeighbors
from sklearn.preprocessing import StandardScaler
```

```python
file_path=r'data/featured_Spotify_track_info.csv'
file_path2=r'data\CLEANED_featured_Spotify_artist_info.csv'
track_data=pd.read_csv(file_path)
artist_data=pd.read_csv(file_path2)
```

```python
print("artist veri seti boyutu: " ,artist_data.shape)
print("müzik veri seti boyutu: " ,track_data.shape)
```

# veri setini inceledik. eksik değerleri mean fonksiyonu ile doldurduk. iki farklı veri setini sanatciId sütunu üzerinden birleştirdik

```python
track_data.head(10)
```

```python
artist_data.head(10)
```

```python
sarkilar = track_data[["ids","artists","names","playlists_found","popularity","danceability","energy","acousticness","instrumentalness","loudness","speechiness","tempo","valence","duration_ms"]]
sarkicilar = artist_data[["ids","names","genres"]]
```

```python
sarkilar = sarkilar.rename(columns= {'artists' : 'sanatciId'})
sarkicilar = sarkicilar.rename(columns={'ids':'sanatciId'})
```

```python
alldataset = pd.merge(sarkilar,sarkicilar,how="inner",on="sanatciId").drop_duplicates()
```

```python
alldataset.head(10)
```

```python
alldataset.shape
```

```python
alldataset.isnull().sum()
```

```python
eksik_sutunlar=['danceability','energy','acousticness','instrumentalness','loudness','speechiness','tempo','valence']
```

```python
alldataset[eksik_sutunlar]
```

```python
alldataset[eksik_sutunlar]=alldataset[eksik_sutunlar].fillna(alldataset[eksik_sutunlar].mean())
```

# sayısal ve hedef değerlerimiz üzerinde standardizasyon işlemlerini gerçekleştirdik. knn algoritmasını en yakın 5 komşu değerini alacak şekilde kurduk ve fit ettik.

```python
scaler = StandardScaler()
sarki_ozellikleri = ['popularity','danceability','energy','acousticness','instrumentalness','loudness','speechiness','tempo','valence','duration_ms']
standardize_veri = scaler.fit_transform(alldataset[sarki_ozellikleri])
print(standardize_veri)
```

```python
alldataset['fullisim'] = alldataset['names_y'] + " - " + alldataset['names_x']
```

```python
knn = NearestNeighbors(n_neighbors=5, metric='cosine')
knn.fit(standardize_veri)
distances, indices = knn.kneighbors(standardize_veri)
```

# uygulamalı kısım 

```python
# ** Kullanıcıdan şarkı ismi al ve öneri yap **
def sarki_oner():
    sorgu = input("hangi şarkının benzerlerini listelememi istersin canım? :p ").strip().lower()
    
    # Şarkıyı bulma
    mask = alldataset['fullisim'].str.lower() == sorgu 
    if not mask.any():
        print("Üzgünüm, bu şarkıyı bulamadım. Başka şarkı girmeyi dene.")
        return
    
    # **Doğru indeksleme yapıyoruz**
    query_index = alldataset[mask].index[0]  # Şarkının orijinal index'i
    query_iloc = alldataset.index.get_loc(query_index)  # Standardize edilmiş veriye uygun index
    
    # Benzer şarkıları bul
    distances, indices = knn.kneighbors(standardize_veri[query_iloc].reshape(1, -1))
    
    # Önerileri yazdır
    print(f"\n'{sorgu}' şarkısına benzer öneriler:")
    for idx in indices[0][1:]:  # İlk indeks kendisi olduğu için hariç tutuyoruz
        print(alldataset.iloc[idx]['fullisim'])

# Kullanıcı sorgu yapabilir
while True:
    sarki_oner()
    devam = input("\nBaşka bir sorgu yapmak ister misin? (E/H): ").strip().lower()
    if devam != 'e':
        print("Sistemden çıkılıyor...")
        break
```
